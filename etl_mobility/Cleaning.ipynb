```python
import pandas as pd
mobility = pd.read_csv('Google Mobility - County - Daily.csv')
```

```python
mobility.info()
```

```python
mobility.head()
```

```python
mobility.isnull().sum()
```

```python
mobility = mobility.replace('.', None)
```

```python
mobility.isnull().sum()
```

```python
# 1. Seleccionamos solo las columnas de movilidad (ignorando año, mes, día y fips)
columnas_gps = mobility.columns[4:]

# 2. Aplicamos la conversión SOLO a esas columnas y sobreescribimos
mobility[columnas_gps] = mobility[columnas_gps].astype('float64')
```

```python
mobility.info()
```

```python
mobility.to_parquet('mobility_clean.parquet')
```
